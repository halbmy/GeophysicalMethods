```python
import pygimli as pg
import pygimli.physics.traveltime as tt
```

```python
data = tt.load("data/koenigsee.sgt")
```

```python
pg.plt.plot(pg.x(data), pg.y(data), "x-")
```

```python
fig, ax = pg.plt.subplots()
tt.drawFirstPicks(ax, data);
```

```python
ax, cb = tt.showVA(data, usePos=False);
cb.ax.set_title("apparent velocity (m/s)");
```

```python
mgr = tt.TravelTimeManager(data)
mgr.invert(zWeight=0.2, verbose=1)
mgr.showResult()
```

```python
mgr.showResult(cMin=500, cMax=3000, cMap="Spectral_r")
```
