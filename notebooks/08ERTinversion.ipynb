```python
import numpy as np
import matplotlib.pyplot as plt
import pygimli as pg
import pygimli.meshtools as mt
import pygimli.physics.ert as ert
from pygimli.physics import ert
plt.style.use("seaborn-notebook")
```

```python
nel = 41
data = ert.createData(nel, schemeName="pd")
print(data)
```

```python
world = mt.createWorld(start=[-5, -10], end=[nel+5, 0], layers=[-2], worldMarker=True)
cube = mt.createRectangle(start=[20, -6], end=[35, -2], marker=3)
geo = world + cube
# pg.show(geo, markers=True);
mesh = mt.createMesh(geo, area=0.5)
pg.show(mesh, markers=True, showMesh=True);
```

```python
rhomap = [[1, 300], [2, 30], [3, 100]]
pg.show(mesh, rhomap, logScale=True, cMap="Spectral_r");
```

```python
data = ert.simulate(mesh, data, rhomap, noiseLevel=0.03, noiseAbs=1)
ert.show(data);
```

```python
mgr = ert.Manager(data)
mgr.invert(verbose=True)
mgr.showResult(cMin=30, cMax=300);
```

```python
mgr.showFit()
```

```python
mgr.showMisfit(errorWeighted=True)
```

```python
mgr.invert(zWeight=0.3, verbose=True)
mgr.showResult(cMin=30, cMax=300);
```

```python
schemes = ["pp", "pd", "wa", "wb", "slm", "dd", "gr"]
for scheme in schemes:
    data = ert.createData(nel, schemeName=scheme)
    data = ert.simulate(mesh, data, rhomap, noiseLevel=0.03, noiseAbs=0)
    mgr = ert.Manager(data)
    mgr.invert()
    ax, cb = mgr.showResult(cMin=30, cMax=300);
    ax.set_title(scheme)
```
